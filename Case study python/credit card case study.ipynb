```python
# Check for missing or null values in the dataset. Handle these appropriately.

print(df.isnull().sum())
```

```python
 # fill the missing value

df['Merchant'].fillna('unknown',inplace=True)
df['Payment_Mode'].fillna('No_idea',inplace=True)
df['Transaction_Status'].fillna('Pending',inplace=True)
df['Category'].fillna('Unkown',inplace=True)
df['Transaction_Type'].fillna('Pending',inplace=True)
print(df.isnull().sum())
```

```python
#fill int values

df['Amount']=df['Amount'].fillna(df['Amount'].median())
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
# changing the transation date object to date 

df['Transaction_Date']=pd.to_datetime(df['Transaction_Date'])
print(df.dtypes)
```

```python
# Convert columns to appropriate data types (e.g., Order Date to datetime).

df['Year']= df['Transaction_Date'].dt.year
df['Month']= df['Transaction_Date'].dt.month
df['day']= df['Transaction_Date'].dt.day

print(df.info())
```

```python
# retrive all transation made in january 2024
```

```python
jan_24=df[(df['Year']==2023) & (df['Month']==1)]
print(jan_24.count())
```

```python
# find trasaction where amount > 3000 AND TRAacation type in 2023

high_transaction =df[(df['Amount']>3000) & (df['Transaction_Type']=="Online")]
print(high_transaction.head())
jan_24.to_csv("jan_24.csv",index=False)
```

```python
# approved transactions

approved_transactions = df[df['Transaction_Status'] == 'Approved']
approved_transactions.head()
```

```python
# create a new clumn discounted amount , assuming a 5% discount on all transaction above $4500

df['Discounted_Amount'] = df['Amount'].apply(lambda x : x*0.80 if x > 4500 else x)
df[['Amount','Discounted_Amount']].head()
```

```python
df['Amount_Category'] = pd.cut(df['Amount'],
                               bins=[-float('inf'),2000,4500,float('inf')],
                               labels=['low','Medium','High'])
df[['Amount_Category','Amount']].head()
```

```python
missing_value= df['Merchant'].isnull().mean()
print(missing_value)

if missing_value > 0.3:
    df.drop(columns=['Merchant'],inplace=True)
    print("Delete done..")
else:
    print("no missing value")
```

```python
# dentify the top 5 products based on total sales revenue.
total_amount_per_category= df.groupby('Category')['Amount'].sum().reset_index()
total_amount_per_category
```

```python
# declined trasaction per category 

declined_transaction_per_category= df[df['Transaction_Status']=='Declined'].groupby('Payment_Mode').size().reset_index(name='declined_count')
declined_transaction_per_category
```

```python
location_transaction= df.groupby('Location')['Amount'].mean().reset_index()
location_transaction
```

```python
print(df.columns)
top_merchants =df ['Merchant'].value_counts().head().reset_index()
top_merchants
```

```python
cust= pd.read_csv("info.csv")
```

```python
merged_df = pd.merge(df,cust,on='Customer_ID',how='inner')
merged_df.head()
```

```python
# find the avg transaction amount per age grp

age_bin = [0,18,30,50,100]
age_labels = ['<18','18-30','31-50','>50']

merged_df['Age_group']=pd.cut(merged_df['Age'],bins=age_bin,labels=age_labels,include_lowest=True)
avg_amount_per_age_group= merged_df.groupby('Age_group')['Amount'].mean().reset_index()
avg_amount_per_age_group
```
